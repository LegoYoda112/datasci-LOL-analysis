```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
gold = pd.read_csv("data/gold.csv")
gold.head()
```

```python
matchinfo = pd.read_csv("data/matchinfo.csv")
matchinfo.head()
```

```python
kills = pd.read_csv("data/kills.csv")
kills.head()
```

```python
match_address = 'http://matchhistory.na.leagueoflegends.com/en/#match-details/TRLH1/1001730150?gameHash=b516729f5838c5e9'

match_kills = kills.loc[kills['Address'] == match_address]

match = matchinfo.loc[matchinfo['Address'] == match_address]
bResult = match['bResult']

print(bResult)

kill_x = match_kills.iloc[:, 9].values.astype(float)
kill_y = match_kills.iloc[:, 10].values.astype(float)
kill_t = match_kills.iloc[:, 2].values.astype(float)

plt.figure(figsize=(10,10))
plt.xlim(0, 16000)
plt.ylim(0, 16000)

plt.scatter(kill_x, kill_y, c=kill_t, cmap=plt.cm.coolwarm)
plt.show()
```

```python
fig, axs = plt.subplots(2, figsize=(15,15))

for match_num in range(8000, 9000):
    match_gold_row = gold.iloc[match_num]
    
    match = matchinfo.loc[matchinfo['Address'] == match_gold_row['Address']]
    
    bResult = match['bResult']
    
    match_gold = match_gold_row[2:]
    
    x = range(0, len(match_gold))
    
    if bResult.values[0]:
        axs[0].plot(x,match_gold, "b", linewidth=1, alpha = 0.4)
    else:
        axs[0].plot(x,match_gold, "r", linewidth=1, alpha = 0.4)
    
plt.show()
```
